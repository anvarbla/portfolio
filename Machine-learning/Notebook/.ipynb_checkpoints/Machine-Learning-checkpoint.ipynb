## **Hypotesis test for categorical columns**

**Airplane model**

```python
m1=df[df["Model"]=="Bombardier CRJ200"]["price"]
m2=df[df["Model"]=="Airbus A320"]["price"]
m3=df[df["Model"]=="Boeing 737"]["price"]
m4=df[df["Model"]=="Cessna 172"]["price"]
m5=df[df["Model"]=="Airbus A350"]["price"]
m6=df[df["Model"]=="Boeing 777"]["price"]
```

```python
#Set the hypothesis
#Ho: mean prices for each airplane model are the same
#H1: mean prices for each airplane model are different

#H0: mu df_m1 = mu df_m2 = mu df_m3 = mu df_m4 = mu df_m5 =mu df_m6
#H1: mu df_m1 != mu df_m2 != mu df_m3 != mu df_m4 != mu df_m5 != mu df_m6

#Lets choose significance level of 10%
alpha = 0.05

f_statistic, p_value = st.f_oneway(m1, m2, m3,m4,m5)

print(f"Statistic_F: {f_statistic}")
print(f"P_value: {p_value}")


if p_value < alpha:
    print("We reject the null hypothesis (H0). There is sufficient evidence to state that at least one mean is different.")
else:
    print("We do not reject the null hypothesis (H0). There is not enough evidence to state that the means are different.")
```

**Airplane engine type**

```python
e1=df["engine_type"]=="Turbofan"
e2=df["engine_type"]=="Piston"
df_e1=df[e1]
df_e2=df[e2]
df_e1=df_e1["price"]
df_e2=df_e2["price"]
```

```python
#Hipotesis test for engine
#Set the hypothesis

#Ho: mean prices for each airplane engine type are the same
#H1: mean prices for each airplane engine type are different

#H0: mu df_e1 = mu df_e2 = mu df_e3 = mu df_e4 = mu df_e5 =mu df_e6
#H1: mu df_e1 != mu df_e2 != mu df_e3 != mu df_e4 != mu df_e5 != mu df_e6

#significance level = 0.05
f_statistic, p_value = st.ttest_ind(df_e1,df_e2, equal_var=False, alternative="two-sided")

print(f"Statistic_F: {f_statistic}")
print(f"P_value: {p_value}")


if p_value < alpha:
    print("We reject the null hypothesis (H0). There is sufficient evidence to state that at least one mean is different.")
else:
    print("We do not reject the null hypothesis (H0). There is not enough evidence to state that the means are different.")
```

**Airplane Region**

```python
r1=df["sales_region"]=="Asia"
r2=df["sales_region"]=="Europe"
r3=df["sales_region"]=="Australia"
r4=df["sales_region"]=="South America"
r5=df["sales_region"]=="Africa"
r6=df["sales_region"]=="North America"

df_r1=df[r1]
df_r2=df[r2]
df_r3=df[r3]
df_r4=df[r4]
df_r5=df[r5]
df_r6=df[r6]

df_r1=pd.DataFrame(df_r1["price"])
df_r2=pd.DataFrame(df_r2["price"])
df_r3=pd.DataFrame(df_r2["price"])
df_r4=pd.DataFrame(df_r4["price"])
df_r5=pd.DataFrame(df_r5["price"])
df_r6=pd.DataFrame(df_r6["price"])
```

```python
#Set the hypothesis
#Ho: mean prices for each airplane region manufacturing are the same
#H1: mean prices for each airplane region manufacturing are different

#H0: mu df_r1 = mu df_r2 = mu df_r3 = mu df_r4 = mu df_r5 =mu df_r6
#H1: mu df_r1 != mu df_r2 != mu df_r3 != mu df_r4 != mu df_r5 != mu df_r6

#Lets choose significance level of 5%
alpha = 0.05

f_statistic, p_value = st.f_oneway(df_r1, df_r2, df_r3, df_r4 , df_r5)

print(f"Statistic_F: {f_statistic}")
print(f"P_value: {p_value}")


if p_value < alpha:
    print("We reject the null hypothesis (H0). There is sufficient evidence to state that at least one mean is different.")
else:
    print("We do not reject the null hypothesis (H0). There is not enough evidence to state that the means are different.")
```

## **Bayesian search**

```python
def objective(trial, confidence_level, folds):

    # First, we define the grid with values to consider when train several possible combinations.
    # Now we specify a range/list of values to try for each hyper-parameter, and we let optuna to decide which
    # combination to try.
    max_depth = trial.suggest_int("max_depth", 2, 150)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 80)
    max_leaf_nodes = trial.suggest_int("max_leaf_nodes", 150, 1000)
    max_features = trial.suggest_categorical("max_features", ["sqrt", "log2"])

    dt = DecisionTreeRegressor(random_state=123,
                               max_depth=max_depth,
                               min_samples_split=min_samples_split,
                               max_leaf_nodes=max_leaf_nodes,
                               max_features=max_features)

    # Here the parameter "cv" specifies the number of folds K
    scores = cross_val_score(dt, X_train, Y_train, cv=folds) # The scores provided will be the R2 on each hold out fold
    mean_score = np.mean(scores)
    sem = np.std(scores, ddof=1) / np.sqrt(folds)

    tc = st.t.ppf(1-((1-confidence_level)/2), df=folds-1)
    lower_bound = mean_score - ( tc * sem )
    upper_bound = mean_score + ( tc * sem )

    # Here, we're storing confidence interval for each trial. It's not possible for the objective function to return
    # multiple values as Optuna uses the only returned value to find the best combination of hyperparameters.
    trial.set_user_attr("CV_score_summary", [round(lower_bound,4), round(np.mean(scores),4), round(upper_bound,4)])

    return np.mean(scores)
```

```python
import optuna
import optuna.visualization as vis
import time

confidence_level = 0.95
folds = 10

start_time = time.time()
study = optuna.create_study(direction="maximize") # We want to have the maximum values for the R2 scores
study.optimize(lambda trial: objective(trial, confidence_level, folds), n_trials=45)
end_time = time.time()

print("\n")
print(f"Time taken to find the best combination of hyperparameters among the given ones: {end_time - start_time: .4f} seconds")
print("\n")
print("The best combination of hyperparameters found was: ", study.best_params)
print(f"The best R2 found was: {study.best_value: .4f}")
```
